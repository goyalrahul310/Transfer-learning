```python
paths = path[1:4]
```

```python
image_data = []
label_data = []
label_dict = {"dogs" : 0,"horses" : 1,"humans" : 2}
```

```python
for p in paths:
    print(p)
    path2 = os.listdir('C:/Users/Rahul/Desktop/coding blocks(rahul)/image data generator/' + folder + "/" +p)
    for p2 in path2:
        imag = image.load_img('C:/Users/Rahul/Desktop/coding blocks(rahul)/image data generator/' + folder + "/" + p + "/" + p2,target_size = (224,224))
        imgtoarray = image.img_to_array(imag)
        image_data.append(imgtoarray)
        label_data.append(label_dict[p])
```

```python
import random
combined = list(zip(image_data,label_data))
random.shuffle(combined)
image_data[:],label_data[:] = zip(*combined)
```

```python
X_train = np.array(image_data)
Y_train_ = np.array(label_data)
```

```python
from keras.utils import np_utils
Y_train = np_utils.to_categorical(Y_train_)
```

```python
print(X_train.shape,Y_train.shape)
```

```python
model = ResNet50(include_top = False, weights='imagenet',input_shape = (224,224,3))
```

```python
av1 = GlobalAveragePooling2D()(model.output)
fc1 = Dense(256,activation='relu')(av1)
d1 = Dropout(0.5)(fc1)
fc2 = Dense(3,activation='softmax')(d1)

model_new = Model(inputs=model.input, outputs= fc2)
model_new.summary()
```

```python
adam = Adam(lr=0.00003)
model_new.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
```

```python
for ix in range(len(model_new.layers)):
    print(ix, model_new.layers[ix])
```

```python
for ix in range(169):
    model_new.layers[ix].trainable = False
    
model_new.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
model_new.summary()
```

```python
hist = model_new.fit(X_train,Y_train,
                    shuffle = True,
                    batch_size = 16,
                    epochs = 5,
                    validation_split=0.20
                    )
```

```python
img_path = 'elephant.jpg'
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

preds = model.predict(x)
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
print('Predicted:', decode_predictions(preds, top=3)[0])
```
